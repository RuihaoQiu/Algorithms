```python
import numpy as np
import tensorflow as tf
import sklearn.datasets
from sklearn.preprocessing import OneHotEncoder
```

```python
## NN classifier
np.random.seed(0)
num_labels = 2


X, y = sklearn.datasets.make_moons(300, noise=0.20)
X = np.float32(X)

enc = OneHotEncoder()
y_onehot = enc.fit_transform(np.reshape(y, (-1, 1))).toarray()

X_tr, X_va, X_te = X[:200], X[200:250], X[250:]
y_tr, y_va, y_te = y_onehot[:200], y_onehot[200:250], y_onehot[250:]

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
XX = np.c_[xx.ravel(), yy.ravel()].astype(np.float32)
```

```python
batch_size = 24
graph = tf.Graph()
with graph.as_default():

  tf_train_dataset = tf.placeholder(tf.float32,
                                    shape=(batch_size, X_tr.shape[1]))
  tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
  tf_valid_dataset = tf.constant(X_va)
  tf_test_dataset = tf.constant(X_te)
  tf_plot = tf.constant(XX)

  num_neurons = 16
  hidden_weights = tf.Variable(tf.truncated_normal([X_tr.shape[1], num_neurons]))
  hidden_biases = tf.Variable(tf.zeros([num_neurons]))
  hidden_layer = tf.nn.relu(tf.matmul(tf_train_dataset, hidden_weights) + hidden_biases)

  weights = tf.Variable(tf.truncated_normal([num_neurons, num_labels]))
  biases = tf.Variable(tf.zeros([num_labels]))
  logits = tf.matmul(hidden_layer, weights) + biases
  
  loss = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits(labels=tf_train_labels, logits=logits))
  optimizer = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
  
  train_prediction = tf.nn.softmax(logits)
    
  valid_relu = tf.nn.relu(tf.matmul(tf_valid_dataset, hidden_weights) + hidden_biases)
  valid_prediction = tf.nn.softmax( tf.matmul(valid_relu, weights) + biases) 
  test_relu = tf.nn.relu( tf.matmul( tf_test_dataset, hidden_weights) + hidden_biases)
  test_prediction = tf.nn.softmax(tf.matmul(test_relu, weights) + biases)

  plot_relu = tf.nn.relu( tf.matmul(tf_plot, hidden_weights) + hidden_biases)
  plot_prediction = tf.nn.softmax(tf.matmul(plot_relu, weights) + biases)
```

```python
num_steps = 5001

def accuracy(predictions, labels):
  return (100.0 * np.sum(np.argmax(predictions, 1) == np.argmax(labels, 1))
          / predictions.shape[0])

with tf.Session(graph=graph) as session:
  tf.global_variables_initializer().run()
  for step in range(num_steps):
    offset = (step * batch_size) % (X_tr.shape[0] - batch_size)
    batch_data = X_tr[offset:(offset + batch_size), :]
    batch_labels = y_tr[offset:(offset + batch_size), :]
    feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
    _, l, predictions = session.run([optimizer, loss, train_prediction], feed_dict=feed_dict)
    if (step % 100 == 0):
      print('Loss at step %d: %f' % (step, l))
      print('Training accuracy: %.1f%%' % accuracy(predictions, batch_labels))
      print('Validation accuracy: %.1f%%' % accuracy(valid_prediction.eval(), y_va))
  print('Test accuracy: %.1f%%' % accuracy(test_prediction.eval(), y_te))
  Z = np.argmax(plot_prediction.eval(), axis=1)
```

```python
## NN regressor
import random

X = np.array([[i*1.0/100 for i in range(100)]]).T.astype(np.float32)
y = np.array([[i**2/10000 + random.uniform(0, 1)*0.1 for i in range(100)]]).T.astype(np.float32)
XX = np.array([[(i+0.5)/100 for i in range(100)]]).T.astype(np.float32)

num_labels = 1
batch_size = 20
graph = tf.Graph()
with graph.as_default():

  tf_train_dataset = tf.placeholder(tf.float32,
                                    shape=(batch_size, X.shape[1]))
  tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
  tf_plot = tf.constant(XX)

  num_neurons = 10
  hidden_weights = tf.Variable(tf.truncated_normal([X.shape[1], num_neurons]))
  hidden_biases = tf.Variable(tf.zeros([num_neurons]))
  hidden_layer = tf.nn.relu(tf.matmul(tf_train_dataset, hidden_weights) + hidden_biases)

  hidden_weights2 = tf.Variable(tf.truncated_normal([num_neurons, num_neurons]))
  hidden_biases2 = tf.Variable(tf.zeros([num_neurons]))
  hidden_layer2 = tf.nn.relu(tf.matmul(hidden_layer, hidden_weights2) + hidden_biases2)

  weights = tf.Variable(tf.truncated_normal([num_neurons, num_labels]))
  biases = tf.Variable(tf.zeros([num_labels]))
  outputs = tf.matmul(hidden_layer2, weights) + biases
  
  loss = tf.reduce_mean(tf.square(outputs - tf_train_labels))
  optimizer = tf.train.GradientDescentOptimizer(0.005).minimize(loss)

  plot_relu = tf.nn.relu(tf.matmul(tf_plot, hidden_weights) + hidden_biases)
  plot_relu2 = tf.nn.relu(tf.matmul(plot_relu, hidden_weights2) + hidden_biases2)
  plot_prediction = tf.matmul(plot_relu2, weights) + biases
```

```python
num_steps = 10001

with tf.Session(graph=graph) as session:
  tf.global_variables_initializer().run()
  for step in range(num_steps):
    offset = (step * batch_size) % (X.shape[0] - batch_size)
    batch_data = X[offset:(offset + batch_size), :]
    batch_labels = y[offset:(offset + batch_size), :]
    feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
    _, l, predictions = session.run([optimizer, loss, outputs], feed_dict=feed_dict)
    if (step % 500 == 0):
        print('Loss at step %d: %f' % (step, l))
  y_pred = plot_prediction.eval()
```
